```python
import os
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import medfilt
import pandas as pd

# Optional: for notebook playback
# from IPython.display import Audio

# Define paths
audio_base_path = "audio_dataset"
hc_audio_path = os.path.join(audio_base_path, "HC_AH")
pd_audio_path = os.path.join(audio_base_path, "PD_AH")
demographics_path = os.path.join(audio_base_path, "Demographics_age_sex.xlsx")

# Load demographics
demographics_df = pd.read_excel(demographics_path)
print("\nDemographics Summary:")
print(demographics_df.describe())

# Load audio data
def load_audio_data(folder_path, max_files=5):
    audio_data = []
    file_names = []

    for file in os.listdir(folder_path):
        if file.endswith('.wav'):
            file_path = os.path.join(folder_path, file)
            try:
                print(f"Loading: {file}")
                audio, sr = librosa.load(file_path, sr=None)
                if len(audio) == 0:
                    print(f"Empty file skipped: {file}")
                    continue
                audio_data.append((audio, sr))
                file_names.append(file)
                if len(audio_data) >= max_files:
                    break
            except Exception as e:
                print(f"Error loading {file}: {e}")
    return audio_data, file_names

hc_audio, hc_files = load_audio_data(hc_audio_path)
pd_audio, pd_files = load_audio_data(pd_audio_path)
print(f"Loaded {len(hc_audio)} HC and {len(pd_audio)} PD audio samples")

# Median filter
def apply_median_filter(signal, kernel_size=5):
    return medfilt(signal, kernel_size)

# Visualization: Original vs Filtered
sample_audio, sr = hc_audio[0]
filtered_audio = apply_median_filter(sample_audio, kernel_size=5)

plt.figure(figsize=(15, 6))
plt.subplot(2, 1, 1)
plt.title('Original Waveform')

plt.subplot(2, 1, 2)
plt.title('Filtered Waveform')

plt.tight_layout()
plt.show()

# Spectrograms
plt.figure(figsize=(15, 8))
plt.subplot(2, 1, 1)
D = librosa.amplitude_to_db(np.abs(librosa.stft(sample_audio)), ref=np.max)
librosa.display.specshow(D, y_axis='hz', x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Original Spectrogram')

plt.subplot(2, 1, 2)
D_filtered = librosa.amplitude_to_db(np.abs(librosa.stft(filtered_audio)), ref=np.max)
librosa.display.specshow(D_filtered, y_axis='hz', x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Filtered Spectrogram')
plt.tight_layout()
plt.show()

# Feature extraction
def extract_audio_features(audio, sr):
    # Pre-process audio signal
    filtered = apply_median_filter(audio, kernel_size=5)
    normalized = librosa.util.normalize(filtered)
    
    # Extract MFCC features with standardized parameters
    n_mfcc = 13
    hop_length = int(sr * 0.010)  # 10ms hop
    n_fft = int(sr * 0.025)      # 25ms window
    mfccs = librosa.feature.mfcc(
        y=normalized, 
        sr=sr, 
        n_mfcc=n_mfcc,
        hop_length=hop_length,
        n_fft=n_fft
    )
    
    # Extract spectral features
    spectral_centroids = librosa.feature.spectral_centroid(
        y=normalized, 
        sr=sr,
        hop_length=hop_length,
        n_fft=n_fft
    )[0]
    
    # Extract zero crossing rate with consistent frame size
    zero_crossing_rate = librosa.feature.zero_crossing_rate(
        normalized,
        hop_length=hop_length
    )[0]
    
    # Calculate statistics with standardized handling
    features = {
        **{f'mfcc_mean_{i+1}': float(np.mean(mfccs[i])) for i in range(n_mfcc)},
        **{f'mfcc_std_{i+1}': float(np.std(mfccs[i])) for i in range(n_mfcc)},
        'spectral_centroid_mean': float(np.mean(spectral_centroids)),
        'spectral_centroid_std': float(np.std(spectral_centroids)),
        'zero_crossing_rate_mean': float(np.mean(zero_crossing_rate)),
        'zero_crossing_rate_std': float(np.std(zero_crossing_rate))
    }
    
    return features

hc_features = [extract_audio_features(audio, sr) for audio, sr in hc_audio]
pd_features = [extract_audio_features(audio, sr) for audio, sr in pd_audio]
print("Feature extraction done.")

# Convert features to DataFrame
def features_to_dataframe(features_list, label):
    df = pd.DataFrame(features_list)
    df['label'] = label
    return df

hc_df = features_to_dataframe(hc_features, 'HC')
pd_df = features_to_dataframe(pd_features, 'PD')
combined_df = pd.concat([hc_df, pd_df], ignore_index=True)

print("\nCombined Feature DataFrame:")
print(combined_df.head())

# Save to CSV
combined_df.to_csv("audio_features_extracted.csv", index=False)
print("Features saved to 'audio_features_extracted.csv'")

# Plot: Compare first MFCC & ZCR
plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 1)
plt.boxplot([
    hc_df['mfcc_mean_1'],
    pd_df['mfcc_mean_1']
], labels=['HC', 'PD'])
plt.title('MFCC Mean #1 Distribution')
plt.ylabel('MFCC Mean Value')

plt.subplot(1, 2, 2)
plt.boxplot([
    hc_df['zero_crossing_rate_mean'],
    pd_df['zero_crossing_rate_mean']
], labels=['HC', 'PD'])
plt.title('Zero Crossing Rate Mean Distribution')
plt.ylabel('ZCR')
plt.tight_layout()
plt.show()
```

```python
import os
import librosa
import librosa.display
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import medfilt, hilbert
from scipy.stats import kurtosis

# -------------------------------
# Feature Extraction Functions
# -------------------------------

def extract_mkmfcc(y, sr, n_mfcc=13, pre_emphasis=0.97):
    y_preemph = np.append(y[0], y[1:] - pre_emphasis * y[:-1])
    mfcc = librosa.feature.mfcc(y=y_preemph, sr=sr, n_mfcc=n_mfcc)
    mfcc_norm = (mfcc - np.mean(mfcc, axis=1, keepdims=True)) / (np.std(mfcc, axis=1, keepdims=True) + 1e-8)
    energy = np.sum(y_preemph ** 2) / len(y_preemph)
    delta = librosa.feature.delta(mfcc)
    mkmfcc_feat = np.concatenate([np.mean(mfcc_norm, axis=1), np.std(mfcc_norm, axis=1), [energy], np.mean(delta, axis=1)])
    return mkmfcc_feat

def extract_bfcc(y, sr, n_bfcc=13):
    S = np.abs(librosa.stft(y))
    mel_basis = librosa.filters.mel(sr=sr, n_fft=2048, n_mels=n_bfcc)
    mel_S = np.dot(mel_basis, S)
    bfcc = np.log(mel_S + 1e-8)
    return np.mean(bfcc, axis=1)

def tonal_power_ratio(y, sr):
    S = np.abs(librosa.stft(y))
    tonal = np.sum(S[:10, :])
    total = np.sum(S)
    return tonal / (total + 1e-8)

def pitch_chroma(y, sr):
    chroma = librosa.feature.chroma_stft(y=y, sr=sr)
    return np.mean(chroma, axis=1)

def fluctuation_index(y, frame_length=1024):
    analytic = hilbert(y)
    imf = np.abs(analytic)
    diff = np.abs(np.diff(imf[:frame_length]))
    return np.sum(diff) / frame_length

def mfcc_features(y, sr, n_mfcc=13):
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
    return np.mean(mfcc, axis=1)

def mean_square_energy(y):
    return np.sqrt(np.mean(y ** 2))

def spectral_flux(y, sr):
    S = np.abs(librosa.stft(y))
    flux = np.sqrt(np.sum(np.diff(S, axis=1) ** 2, axis=0))
    return np.mean(flux)

def spectral_spread(y, sr):
    S, _ = librosa.magphase(librosa.stft(y))
    centroid = librosa.feature.spectral_centroid(S=S, sr=sr)
    spread = np.sqrt(np.sum(((np.arange(S.shape[0])[:, None] - centroid) ** 2) * S, axis=0) / (np.sum(S, axis=0) + 1e-8))
    return np.mean(spread)

def spectral_centroid(y, sr):
    centroid = librosa.feature.spectral_centroid(y=y, sr=sr)
    return np.mean(centroid)

def spectral_kurtosis(y, sr):
    S = np.abs(librosa.stft(y))
    kurt = kurtosis(S, axis=0, fisher=False, bias=False)
    return np.mean(kurt)

def spectral_rolloff(y, sr, roll_percent=0.95):
    rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr, roll_percent=roll_percent)
    return np.mean(rolloff)

def spectral_bandwidth(y, sr):
    bandwidth = librosa.feature.spectral_bandwidth(y=y, sr=sr)
    return np.mean(bandwidth)

def spectral_flatness(y):
    flatness = librosa.feature.spectral_flatness(y=y)
    return np.mean(flatness)

def extract_all_voice_features(y, sr):
    try:
        S1 = extract_mkmfcc(y, sr)
        S2 = extract_bfcc(y, sr)
        S3 = tonal_power_ratio(y, sr)
        S4 = pitch_chroma(y, sr)
        S5 = fluctuation_index(y)
        S6 = mfcc_features(y, sr)
        S7 = mean_square_energy(y)
        S8 = spectral_flux(y, sr)
        S9 = spectral_spread(y, sr)
        S10 = spectral_centroid(y, sr)
        S11 = spectral_kurtosis(y, sr)
        S12 = spectral_rolloff(y, sr)
        S13 = spectral_bandwidth(y, sr)
        S14 = spectral_flatness(y)

        features = np.concatenate([
            S1.flatten(),
            S2.flatten(),
            [S3],
            S4.flatten(),
            [S5],
            S6.flatten(),
            [S7, S8, S9, S10, S11, S12, S13, S14]
        ])
        return features

    except Exception as e:
        print(f"Error in extract_all_voice_features: {e}")
        raise

# -------------------------------
# Load & Extract for Dataset
# -------------------------------

def load_audio_data(folder_path, max_files=5):
    audio_data = []
    file_names = []
    for file in os.listdir(folder_path):
        if file.endswith('.wav'):
            try:
                path = os.path.join(folder_path, file)
                y, sr = librosa.load(path, sr=None)
                if len(y) == 0:
                    print(f"Empty file: {file}")
                    continue
                audio_data.append((y, sr))
                file_names.append(file)
                if len(audio_data) >= max_files:
                    break
            except Exception as e:
                print(f"Failed to load {file}: {e}")
    return audio_data, file_names

# -------------------------------
# Example Execution
# -------------------------------

audio_base_path = "audio_dataset"
hc_audio_path = os.path.join(audio_base_path, "HC_AH")
pd_audio_path = os.path.join(audio_base_path, "PD_AH")

hc_audio, hc_files = load_audio_data(hc_audio_path)
pd_audio, pd_files = load_audio_data(pd_audio_path)

hc_voice_features = [extract_all_voice_features(y, sr) for y, sr in hc_audio]
pd_voice_features = [extract_all_voice_features(y, sr) for y, sr in pd_audio]

print(f"Extracted features for {len(hc_voice_features)} HC and {len(pd_voice_features)} PD samples.")
```

```python
import numpy as np

def canberra_distance(a, b):
    return np.sum(np.abs(a - b) / (np.abs(a) + np.abs(b) + 1e-8))

def fitness(population, data, labels):
    # Example: use mean Canberra distance to class means as fitness (lower is better)
    class0 = data[labels == 0]
    class1 = data[labels == 1]
    mean0 = np.mean(class0, axis=0)
    mean1 = np.mean(class1, axis=0)
    fit = []
    for p in population:
        sel = np.where(p == 1)[0]
        if len(sel) == 0:
            fit.append(np.inf)
        else:
            d0 = np.mean([canberra_distance(x[sel], mean1[sel]) for x in class0])
            d1 = np.mean([canberra_distance(x[sel], mean0[sel]) for x in class1])
            fit.append(d0 + d1)
    return np.array(fit)

def lho_feature_selection(X, y, n_ladybugs=20, max_iter=30, mutation_rate=0.2, early_stop_patience=5):
    # Input validation
    if len(X) != len(y):
        raise ValueError(f"X and y must have same length. Got X: {len(X)}, y: {len(y)}")
    if not isinstance(X, np.ndarray) or not isinstance(y, np.ndarray):
        X = np.array(X)
        y = np.array(y)
    
    # Initialize variables with clear types
    n_features: int = X.shape[1]
    population: np.ndarray = np.random.randint(0, 2, (n_ladybugs, n_features))
    best_fitness: float = np.inf
    best_solution: np.ndarray = None
    patience_counter: int = 0
    last_best_fitness: float = np.inf
    
    # LHO parameters
    omega1: float = 0.5  # Weight for best solution
    omega2: float = 0.5  # Weight for neighboring solution
    
    for iteration in range(max_iter):
        try:
            # Calculate fitness with explicit typing
            fitness_values: np.ndarray = fitness(population, X, y)
            current_best_idx: int = np.argmin(fitness_values)
            current_best_fitness: float = fitness_values[current_best_idx]
            
            # Update best solution if improved
            if current_best_fitness < best_fitness:
                best_fitness = current_best_fitness
                best_solution = population[current_best_idx].copy()
                
                # Reset patience if improved
                if current_best_fitness < last_best_fitness:
                    patience_counter = 0
                    last_best_fitness = current_best_fitness
            else:
                patience_counter += 1
            
            # Early stopping check
            if patience_counter >= early_stop_patience:
                print(f"Early stopping at iteration {iteration} due to no improvement")
                break
                
            # Create new population with type safety
            new_population = []
            for b in range(n_ladybugs):
                # Generate random values for movement
                rand1: float = np.random.random()
                rand2: float = np.random.random()
                m: int = np.random.randint(n_ladybugs)
                
                # Calculate new position
                pm: np.ndarray = population[m]
                pm1: np.ndarray = population[(m+1)%n_ladybugs]
                new_pos: np.ndarray = population[b] + rand1 * (pm - population[b]) + rand2 * (omega1 * best_solution - omega2 * pm1)
                
                # Apply mutation with explicit mask
                mutation_mask: np.ndarray = np.random.random(n_features) < mutation_rate
                new_pos[mutation_mask] = 1 - new_pos[mutation_mask]
                
                # Ensure binary output
                new_population.append((new_pos > 0.5).astype(np.int32))
            
            # Update population
            population = np.array(new_population)
            
            # Print progress every 5 iterations
            if iteration % 5 == 0:
                print(f"Iteration {iteration}: Best fitness = {best_fitness:.4f}")
                
        except Exception as e:
            print(f"Error in iteration {iteration}: {str(e)}")
            break
    
    # Validate final solution
    if best_solution is None:
        raise ValueError("No solution found by LHO")
        
    # Get selected features
    selected_features = np.where(best_solution == 1)[0]
    print(f"Selected {len(selected_features)} features")
    
    return selected_features
```

```python
import torch

import torch.nn as nn
import torch.nn.functional as F

# ZFNet implementation (input shape: [batch, 3, 224, 224])
class ZFNet(nn.Module):
    def __init__(self, num_classes=2):
        super(ZFNet, self).__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 96, kernel_size=7, stride=2, padding=1),  # Conv1
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=3, stride=2),                 # Pool1

            nn.Conv2d(96, 256, kernel_size=5, stride=2, padding=1), # Conv2
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=3, stride=2),                 # Pool2

            nn.Conv2d(256, 384, kernel_size=3, stride=1, padding=1), # Conv3
            nn.ReLU(inplace=True),

            nn.Conv2d(384, 384, kernel_size=3, stride=1, padding=1), # Conv4
            nn.ReLU(inplace=True),

            nn.Conv2d(384, 256, kernel_size=3, stride=1, padding=1), # Conv5
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=3, stride=2)                  # Pool5
        )
        self.classifier = nn.Sequential(
            nn.Dropout(),
            nn.Linear(256 * 6 * 6, 4096),
            nn.ReLU(inplace=True),
            nn.Dropout(),
            nn.Linear(4096, 4096),
            nn.ReLU(inplace=True),
            nn.Linear(4096, num_classes)
        )

    def forward(self, x):
        x = self.features(x)
        x = torch.flatten(x, 1)
        x = self.classifier(x)
        return x

# Example: create ZFNet model
zfnet = ZFNet(num_classes=2)
print(zfnet)
```

```python
def train_and_evaluate_model(model, X_train, X_test, y_train, y_test, class_weights, model_name):
    print(f"\nTraining {model_name}...")
    
    if isinstance(model, xgb.XGBClassifier):
        # For XGBoost, use scale_pos_weight
        scale_pos_weight = class_weights[1] / class_weights[0]
        model.set_params(scale_pos_weight=scale_pos_weight)
        # XGBoost 1.6+ uses callbacks for early stopping
        early_stopping = xgb.callback.EarlyStopping(
            rounds=10,
            metric_name='logloss',
            save_best=True
        )
        model.fit(
            X_train, y_train,
            eval_set=[(X_test, y_test)],
            callbacks=[early_stopping],
            verbose=False
        )
    
    elif isinstance(model, LGBMClassifier):
        # For LightGBM, ensure feature names are set
        if isinstance(X_train, pd.DataFrame):
            feature_names = X_train.columns.tolist()
            model.set_params(feature_name=feature_names)
        
        # Use class weights directly
        model.set_params(class_weight=class_weights)
        eval_set = [(X_test, y_test)]
        model.fit(
            X_train, y_train,
            eval_set=eval_set,
            early_stopping_rounds=10,
            verbose=False
        )
    
    # Make predictions
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:,1]
    
    # Calculate metrics with proper zero division handling
    results = {
        'accuracy': accuracy_score(y_test, y_pred),
        'precision': precision_score(y_test, y_pred, zero_division=0),
        'recall': recall_score(y_test, y_pred, zero_division=0),
        'f1': f1_score(y_test, y_pred, zero_division=0),
        'roc_auc': roc_auc_score(y_test, y_pred_proba)
    }
    
    return results, y_pred, y_pred_proba
```

```python
# Evaluate and visualize results 
def plot_confusion_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(title)
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

# Train and evaluate on image data
print("\nEvaluating models on image data:")
xgb_img_results, xgb_img_pred, xgb_img_proba = train_and_evaluate_model(
    models['XGBoost'], X_img_train, X_img_test, y_img_train, y_img_test, 
    img_weight_dict, "XGBoost (Image)"
)

lgb_img_results, lgb_img_pred, lgb_img_proba = train_and_evaluate_model(
    models['LightGBM'], X_img_train, X_img_test, y_img_train, y_img_test, 
    img_weight_dict, "LightGBM (Image)"
)

# Train and evaluate on audio data
print("\nEvaluating models on audio data:")
xgb_audio_results, xgb_audio_pred, xgb_audio_proba = train_and_evaluate_model(
    models['XGBoost'], X_audio_train, X_audio_test, y_audio_train, y_audio_test, 
    audio_weight_dict, "XGBoost (Audio)"
)

lgb_audio_results, lgb_audio_pred, lgb_audio_proba = train_and_evaluate_model(
    models['LightGBM'], X_audio_train, X_audio_test, y_audio_train, y_audio_test, 
    audio_weight_dict, "LightGBM (Audio)"
)

# Print results
print("\nXGBoost Results:")
print("Image:", xgb_img_results)
print("Audio:", xgb_audio_results)

print("\nLightGBM Results:")
print("Image:", lgb_img_results)
print("Audio:", lgb_audio_results)

# Plot confusion matrices
plot_confusion_matrix(y_img_test, xgb_img_pred, 'XGBoost - Image Data Confusion Matrix')
plot_confusion_matrix(y_img_test, lgb_img_pred, 'LightGBM - Image Data Confusion Matrix')
plot_confusion_matrix(y_audio_test, xgb_audio_pred, 'XGBoost - Audio Data Confusion Matrix')
plot_confusion_matrix(y_audio_test, lgb_audio_pred, 'LightGBM - Audio Data Confusion Matrix')
```
